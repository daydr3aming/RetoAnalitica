```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
fig = plt.figure(figsize=(8, 5))
numeric = df.drop('Species', axis=1)
corr = numeric.corr()
sns.heatmap(data = corr, vmin= -1, vmax = 1, cmap = "OrRd", annot = True)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))
sns.boxplot(x = 'Species', y = 'Sepal.Length', data = df)
plt.title('Sepal Length by Species')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 5))
sns.histplot(data = df, x = 'Sepal.Length', hue = 'Species', bins=10, kde = True, ax = axs[0][0])
sns.histplot(data = df, x = 'Sepal.Width', hue = 'Species', bins=10, kde = True, ax = axs[0][1])
sns.histplot(data = df, x = 'Petal.Length', hue = 'Species', bins=10, kde = True, ax = axs[1][0])
sns.histplot(data = df, x = 'Petal.Width', hue = 'Species', bins=10, kde = True, ax = axs[1][1])
```
